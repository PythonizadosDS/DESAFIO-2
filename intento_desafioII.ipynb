```python
def change_column_order(df, col_name, index):
    cols = df.columns.tolist()
    cols.remove(col_name)
    cols.insert(index, col_name)
    
    return df[cols]
```

```python
df = change_column_order(bill,'Periodo_time', 6)
```

```python
#Mejor ranking alcanzado
df['mejor_rank'] = df.iloc[:,7:].min(axis=1)
```

```python
#total rank por  track
#df.iloc[:,7:].sum(skipna = True)
df['total_rank_tr'] =df.iloc[:,7:].sum(axis = 'columns')
```

```python
df = change_column_order(df,'mejor_rank', 7)
```

```python
df = change_column_order(df,'total_rank_tr', 8)
```

```python
help(pd.melt)
```

```python
id_vars = ["artist.inverted", "track", "time", "genre",  "date.entered", "date.peaked","Periodo_time"]
```

```python
bill_tidy = pd.melt(frame=df,id_vars=id_vars, var_name="week", value_name="rank")
```

```python
bill_tidy.head()
```

```python
def periodo_en_semanas(x):
    semanas = x/7
    return semanas
```

```python
bill_tidy["Periodo_time"] = bill_tidy["Periodo_time"].apply(periodo_en_semanas)
bill_tidy.head()
```

```python

# Formatting 
bill_tidy["week"] = bill_tidy["week"].str.extract('(\d+)', expand=False).astype(float)
```

```python
bill_tidy["rank"] = bill_tidy["rank"].astype(float)
```

```python
bill_f = bill_tidy.dropna()
```

```python
bill_f['date'] = pd.to_datetime(bill_f['date.entered']) + pd.to_timedelta(bill_f['week'], unit='w') - pd.DateOffset(weeks=1)
```

```python
bill_f = bill_f[[  "artist.inverted",  "track",  "time",  "genre", "week",  "rank", "date"]]
df = bill_f.sort_values(ascending=True, by=["artist.inverted","track","week","rank"])

# Assigning the tidy dataset to a variable for future usage
billboard = df
```
